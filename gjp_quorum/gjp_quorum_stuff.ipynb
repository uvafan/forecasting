```python
import pandas as pd

def get_weekly_curve(country, col):
    df = pd.read_csv(f'../data/worldometers/{country}.csv')
    weekly_amts = []
    for week_end in range(len(df)-1, 6, -7):
        weekly_amts.append(df.at[week_end,col] - df.at[week_end-7,col])
    return list(reversed(weekly_amts))

countries = ['france','italy','japan','sk','spain','us', 'india', 'brazil', 'germany']

death_curves = {}
case_curves = {}
for country in countries:
    death_curves[country] = get_weekly_curve(country, 'coronavirus-deaths-linear')
    case_curves[country] = get_weekly_curve(country, 'coronavirus-cases-linear')

death_curves
```

```python
import matplotlib.pyplot as plt
for d in death_curves:
    plt.plot(death_curves[d])
plt.legend(death_curves.keys())
plt.show()
```

```python
for c in death_curves:
    plt.plot(case_curves[c])
plt.legend(death_curves.keys())
plt.show()
```
